## EDA With Red Wine Data

```python
import pandas as pd
dataframe=pd.read_csv('winequality-red.csv')
dataframe.head()
```

```python
## Summary of the dataset
dataframe.info()
```

```python
## descriptive summary of the dataset
dataframe.describe()
```

```python
dataframe.shape
```

```python
## List down all the columns names
dataframe.columns
```

```python
# Finding the unique values of the quality variable
dataframe['quality'].unique()
```

```python
## Missing values in the dataset
dataframe.isnull().sum()
```

```python
## Identifing the Duplicate records
dataframe[dataframe.duplicated()]
```

```python
## Remove the duplicates
New_dataframe = dataframe.copy()
New_dataframe.drop_duplicates(inplace=True)
```

```python
# new dataframe shape after removing the duplicates
New_dataframe.shape
```

```python
# Old dataframe shape 
dataframe.shape
```

Remember to make copy of dataframe if assigning a new variable otherwise all the operations on new dataframes will be done on the original dataframes as well

```python
## Correlation
New_dataframe.corr()
```

Now let's visualise the Correlation Matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(New_dataframe.corr(),annot=True)
```

Now, Visualising the distribution of the dataset

```python
## Visualization
New_dataframe.quality.value_counts().plot(kind='bar')
plt.xlabel("Wine Quality")
plt.ylabel("Count")
plt.show()
```

Conclusion- It is an imbalanced dataset

```python
New_dataframe.head()
```

Now lets visualise all the other features

```python
features = New_dataframe.columns
```

```python
features
```

```python
for feature in features:
    print(f'Histogram for {feature}')
    plt.figure(figsize=(10,6))
    plt.hist(New_dataframe[feature], bins = 20)
    plt.xlabel(f'{feature} values')
    plt.show()
    print('\n')
```

Visualising all features together
